Совместим координаты отслеживаемого объекта с кадрами видео

```python
import matplotlib.pyplot as plt
import pandas as pd
import cv2
```

```python
head = ['x', 'y', 'w', 'h']
df_res = pd.read_csv('/content/results/NCCPython/baseline/basketball/basketball_001.txt', sep=',', names=head,  header=0)
df_res.head(10)
# не на всех кадрах объект был обнаружен, в таких случаях строки имеют вид "0.0"
```

```python
imgs = []
for i in range(1,726):
  name = '0'*(8-len(str(i))) + str(i)
  path_img = f'/content/sequences/basketball/color/{name}.jpg'
  img = plt.imread(path_img)
  #попытка добавить маску объекта на кадр, в случае неудачи в результате будет оригинальный кадр
  try:
    x1 = int(df_res.iloc[i-1:i,:].x.to_list()[0])
    y1 = int(df_res.iloc[i-1:i,:].y.to_list()[0])
    x2 = int(df_res.iloc[i-1:i,:].x.to_list()[0]+df_res.iloc[1:2,:].w.to_list()[0]) 
    y2 = int(df_res.iloc[i-1:i,:].y.to_list()[0]+df_res.iloc[1:2,:].h.to_list()[0]) 
    color_red = (0,0,255)
    img2 = cv2.rectangle(img, (x1,y1),(x2,y2),color_red, thickness=2, lineType=8, shift=0)
  except BaseException:
    img2 = img
  imgs.append(img2)
```

```python
imgs.__len__()
```

```python
!mkdir resWithRectangles
for i in range(len(imgs)):
  cv2.imwrite(f'/content/resWithRectangles/{i}.png', imgs[i])
```
